#### Removing null values

```python
temp_df.dropna()
```

This operation will delete any row with at least a single null value, but it will return a new DataFrame without altering the original one. You could specify inplace=True in this method as well.

```python
temp_df.dropna(axis=1)
#drops entire column having a null value
```

What's with this axis=1 parameter?

It's not immediately obvious where axis comes from and why you need it to be 1 for it to affect columns. To see why, just look at the .shape output:

temp_df.shape

Out: (41188, 9)

As we learned above, this is a tuple that represents the shape of the DataFrame, i.e. 41188 rows and 9 columns. Note that the rows are at index zero of this tuple and columns are at index one of this tuple. This is why axis=1 affects columns. This comes from NumPy, and is a great example of why learning NumPy is worth your time.

Using above methods we can loose lot of data, Imputation

#### Imputation

Imputation is a conventional feature engineering technique used to keep valuable data that have null values.

```python
data = {
    'apples': [3, 2, 0, 1,None,8,3,1,5,None,7,None,8,5,4], 
    'oranges': [0, 3, 7, 2,1,7,None,5,3,8,6,None,5,3,None]
}
purchases = pd.DataFrame(data)
```
